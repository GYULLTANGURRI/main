```python
from hossam import load_data
from pandas import DataFrame, get_dummies, merge
from sklearn.preprocessing import OneHotEncoder
```

```python
origin = load_data("nursing_grades")
origin
```

```python
df = origin.astype({'성별':'category', '병원경력':'category','합격여부':'category'})
df.info()
```

```python
df1= get_dummies(df, columns=['성별'], dtype = 'int')
df1.head()
```

```python
# N-1더미변수 생성
df2 = get_dummies(df, columns = ['성별'], dtype = 'int', drop_first = True)
df.head()
```

```python
df3 = get_dummies( df, columns = ['성별', '병원경력'],dtype = 'int',
                  drop_first = True)
df3.head()
```

```python
x = df['성별'].values.reshape(-1,1)
x
```

```python
x = df[['성별']]
x
```

```python
encoder = OneHotEncoder(sparse_output=False, drop=None)
result = encoder.fit_transform(x)
result
```

```python
# Scikit-Learn을 사용하는 방법(트리 기반 머신러닝 모형에 추천)
new_cols = encoder.get_feature_names_out(['성별'])
new_cols
```

```python
one_hot_df = DataFrame(result, columns = new_cols, index = df.index)
one_hot_df.head()
```

```python
df_copy = df.copy()
df4 = merge(df_copy, one_hot_df, left_index = True, right_index = True)
df4.head()
```

# 온라인 스토어 고객 행동 분석

```python
import pandas as pd
```

```python
customers = load_data("online_store_customers")
customers
```

```python
purchases = load_data("online_store_purchases")
purchases
```

```python
# 1.
merged = pd.merge(customers, purchases, on='user_id', how = 'inner')
merged
```

```python
# 2.
size_mode = merged['size'].mode()[0]

size_mode
```

```python
merged['size'] = merged['size'].fillna(size_mode)
merged
```

```python
# 3.
gender_dummies = pd.get_dummies(merged['gender'], prefix = 'gender')
merged = pd.concat([merged, gender_dummies],axis=1)
merged
```

```python
# 3.
color_dummies = pd.get_dummies(merged['color'], prefix = 'color')
merged = pd.concat([merged, color_dummies], axis = 1)
merged
```
