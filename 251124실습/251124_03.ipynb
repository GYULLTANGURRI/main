```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
from pandas import pivot_table
import seaborn as sb
import numpy as np
```

```python
my_dpi = 200
fpath = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(fpath)
fprop=fm.FontProperties(fname=fpath)
fname=fprop.get_name()


plt.rcParams['font.family'] = fname
plt.rcParams['font.size'] = 6
plt.rcParams['axes.unicode_minus'] = False

```

```python
origin = load_data("penguins")
origin
```

###02.barplot

```python
width_px = 1280
height_px = 720
rows = 1
cols=1
figsize = (width_px /my_dpi, height_px/my_dpi)
fig, ax = plt.subplots(rows,cols,figsize=figsize, dpi=my_dpi)

# 그래프 그리기->seaborn 사용
sb.barplot(data=origin, x="species", y="body_mass_g", hue = "sex", 
           estimator=np.mean, errorbar=("ci", 95),palette="Set2")
           
ax.grid(True)


plt.tight_layout()
plt.savefig("plot.png", dpi=my_dpi*2)
plt.show()
plt.close()
```

###03.countplot(빈도 그래프)

```python
width_px = 1280
height_px = 760
rows = 1
cols=1
figsize = (width_px /my_dpi, height_px/my_dpi)
fig, ax = plt.subplots(rows,cols,figsize=figsize, dpi=my_dpi)

# 그래프 그리기->seaborn 사용
sb.countplot(data=origin, x="species", hue="sex",order=['Chinstrap', 'Gentoo','Adelie'], palette = "Pastel1")

ax.grid(True)


plt.tight_layout()
plt.savefig("plot.png", dpi=my_dpi*2)
plt.show()
plt.close()
```

###04.pointplot

```python
width_px = 1280
height_px = 720
rows = 1
cols=1
figsize = (width_px /my_dpi, height_px/my_dpi)
fig, ax = plt.subplots(rows,cols,figsize=figsize, dpi=my_dpi)

# 그래프 그리기->seaborn 사용
sb.pointplot(data=origin, x="species", y="flipper_length_mm", hue = "sex",estimator = np.mean,
            errorbar=("ci",95), linestyles="--", linewidth=1, markers="o", palette="Dark2")
           
ax.grid(True)

plt.tight_layout()
plt.savefig("plot.png", dpi=my_dpi*2)
plt.show()
plt.close()
```
