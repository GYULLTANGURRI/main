```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
import seaborn as sb
import numpy as np
```

```python
my_dpi = 200
fpath = "./NotoSansKr-Regular.ttf"
fm.fontManager.addfont(fpath)
fprop = fm.FontProperties(fname=fpath)
fname = fprop.get_name()



plt.rcParams['font.family'] = fname
plt.rcParams['font.size'] = 6
plt.rcParams['axes.unicode_minus'] = False
```

```python
origin = load_data('traffic_acc')
origin
```

# 데이터 전처리

```python
df = origin.drop('월', axis=1).groupby('년도').mean()
df
```

# 02.서브플롯의 기본 사용

```python
from matplotlib import pyplot as plt
fig, axes = plt.subplots(2,3, figsize=(10, 6))
```

```python
width_px = 2000
height_px = 1500
rows = 2
cols=3
figsize = (width_px /my_dpi, height_px/my_dpi)

fig, ax = plt.subplots(rows,cols,figsize=figsize, dpi=my_dpi)

# 그래프 그리기->seaborn 사용
fig.suptitle('서브플롯 영역 나누기', fontsize = 14, color= "#2124E7")
           
fig.subplots_adjust(wspace=0.2, hspace=0.2)



plt.tight_layout()
plt.savefig("plot.png", dpi=my_dpi*2)
plt.show()
plt.close()
```

```python
width_px = 2400
height_px = 1500
rows = 2
cols=3
figsize = (width_px /my_dpi, height_px/my_dpi)
fig, ax = plt.subplots(rows,cols,figsize=figsize, dpi=my_dpi)

# 그래프 그리기->seaborn 사용
sb.boxplot(data=df, y = '발생건수', ax = ax[0][0])
sb.histplot(data=df, x='사망자수', bins=5,
            ax=ax[0][1])
sb.kdeplot(data=df, x=df.index, y='발생건수', ax=ax[0][2])
sb.lineplot(data=df, x=df.index, y='발생건수', ax=ax[1][0])
sb.barplot(data=df, x=df.index, y='사망자수', ax=ax[1][2])          
# 그래프 꾸미기
fig.suptitle('년도별 교통사고집계', fontsize = 14, color= "#000000")
fig.subplots_adjust(wspace= 0.2, hspace = 0.2 )
ax[0][0].set_title("교통사고 발생 건수 분포", color = "#000000", fontsize = 8,
                   pad = 8)
ax[0][0].grid()

hist, bins = np.histogram(df['사망자수'], bins = 5)
bins = bins.round().astype("int")
ax[0][1].set_title("교통사고 사망자수 히스토그램", color = "#000000", fontsize = 8, pad = 8)

ax[0][1].set_xticks(bins,bins)
ax[0][1].set_xlabel('구간')
ax[0][1].set_ylabel('집계')
ax[0][1].grid()





plt.tight_layout()
plt.show()
plt.close()
```

# 03.두 개의 y축을 갖는 그래프

```python
# 1) x축 데이터
x= np.arange(10)
x
```

```python
#2)y축 데이터
y1 = np.arange(10)
y1
```

```python
# 3)두 번째 y축 데이터
y2 = x **2
y2
```

2.서브플롯으로 2 개의 y축을 갖는 그래프 구현

```python
width_px = 1280
height_px = 720
rows = 1
cols=1
figsize = (width_px /my_dpi, height_px/my_dpi)
fig, ax1 = plt.subplots(rows,cols,figsize=figsize, dpi=my_dpi)

ax2 = ax1.twinx()
# 그래프 그리기->seaborn 사용
sb.lineplot( x=x,y=y1,color="#000000", ax=ax1)
sb.lineplot( x=x,y=y2,color="#000000", ax=ax2)        


ax1.grid(True)

plt.tight_layout()
plt.show()
plt.close()
```

#04.교통사고 발생건수 사망자수 변화 시각화

```python
width_px = 1280
height_px = 720
rows = 1
cols=1
figsize = (width_px /my_dpi, height_px/my_dpi)
fig, ax1 = plt.subplots(rows,cols,figsize=figsize, dpi=my_dpi)

ax2 = ax1.twinx()
# 그래프 그리기->seaborn 사용
sb.lineplot(data=df, x=df.index,y='발생건수',color="#000000", ax=ax1, label='발생건수')
sb.lineplot(data=df,x=df.index,y= '부상자수',color="#000000", ax=ax2, label='부상자수')        

ax1.set_xlabel('년도')
ax1.set_ylabel('발생건수',color= "#000000" )
ax1.grid(True)
ax1.legend(loc='upper right')
ax2.legend(loc='lower right')

plt.tight_layout()
plt.show()
plt.close()
```
