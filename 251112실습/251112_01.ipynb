```python
import numpy as np
from pandas import Series
```

```python
# 리스트를 활용한 객체 생성
my_array = [100,200,300]
s1 = Series(my_array)
s1
```

```python
# ndarray를 통한 객체 생성
my_array = np.array([100,200,300])
s2 = Series(my_array)
s2
```

```python
# 딕셔너리를 통한 객체 생성
my_array = {"hello":123, "world":456, "nice":789}
s3 = Series(my_array)
s3
```

```python
my_array = Series({"hello":123, "world":456, "nice":789})
my_array.dtype
```

```python
my_array.info()
```

```python
my_series = Series([10,20,30,40])
print("==인덱싱==")
print(my_series[0])
print("==슬라이싱==")
print(my_series[1:3])
```

```python
# 딕셔너리를 통해 생성
my_series = Series({"a":100, "b":200, "c":300, "d":400})

print("==인덱싱==")
print(my_series["a"])


print("==슬라이싱==")
print(my_series["b":"d"])

```

```python
# value 프로퍼티
mylist = list(my_series)
print(type(my_series.values))
my_series.values
```

```python
mylist = list(my_series.values)
mylist
```

```python
my_series = Series({"a":100, "b":200, "c":300, "d":400})
print(type(my_series.index))
my_series.index
```

```python
mylist = list(my_series.index)
mylist
```
