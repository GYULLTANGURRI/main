```python
import numpy as np
```

```python
A = np.array([88,92,76,100,85,81,79,98,85,87])
B = np.array([65,99,78,72,80,91,75,82,89,70])
A_array = len(A)
B_array = len(B)
print("Alpa팀의 도수:", A_array)
print("Brabo팀의 도수:", B_array)

A_sorted = sorted(A)
B_sorted = sorted(A)

print("Alpha팀의 평균:", np.average(A_sorted))
print("Bravo팀의 평균:", np.average(B_sorted))

print("Alpha팀의 표준편차:", np.std(A_sorted))
print("Bravo팀의 표준편차:", np.std(B_sorted))

print("Alpha팀의 최솟값:", np.min(A_sorted))
print("Bravo팀의 최솟값:", np.min(A_sorted ))


Aq_1 = (A_array-1) * 0.25
Bq_1 = (B_array-1) * 0.25
A_i = int(np.floor(Aq_1))
B_i = int(np.floor(Bq_1))
f_A= Aq_1 - A_i
f_B = Bq_1 - B_i
Q1_A = A_sorted[A_array-1] +  f_A * (A_sorted[A_array+1] - A_sorted[A_array])
print("Alpha의 1사분위 수:", Q1_A)

Q1_B = B_sorted[B_array-1] + f_B * (B_sorted[B_array+1] - B_sorted[B_array])
print("Bravo팀의 1사분위수:", Q1_B)


# 중위값
print("Alpha의 중앙값:", np.median(A))
print("Bravo의 중앙값:", np.median(B))

# 3사분위 수
print("Alpha의 3사분위수:", np.quantile(A, 0.75)- np.quantile (A, 0.25))
print("Bravo의 3사분위수:", np.quantile(B, 0.75) - np.quantile(B, 0.25))


# 최댓값
print("Alpha의 최댓값:", np.max(A))
print("Bravo의 최댓값:", np.max(B))


```
