```python
import numpy as np
```

```python

np.random.seed(100)
np.random.normal(75, 5, (3,4))

score = [66.25117263, 76.71340202, 80.76517901, 73.73781982,79.90660393, 77.57109421, 76.10589835, 69.6497833,
         74.05252085, 76.27500722, 72.70986507, 77.17581744]

score = np.array(score)
print(score)

result = score[score >= 80]
print(f"80점 이상 점수:{result}")

print("차원:", score.ndim)
print("형태", score.shape)
print("데이터 타입", score.dtype)


```
