```python
import numpy as np
```

```python
a = np.arange(12)
print("원본 배열:",a)
print("실수형 반환:", a.astype(float))

b = a.reshape(2,6)
print("2행 6열 형태:", b)

c = a.reshape(3,-1)
print("자동 계산된 형태(3행 4열):",c)

d = a.flatten()
print("1차원 배열로 축소:",d)










```
