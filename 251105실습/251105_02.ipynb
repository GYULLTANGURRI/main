#03. 카카오 검색 데이터 수집 (2)

```python
# 라이브러리 참조하기
import requests
import datetime as dt
from pandas import DataFrame
```

```python
# 요청 정보 확인
url = "https://dapi.kakao.com/v3/search/book"

query = "python"
page = 1
size = 50
key = "375a1f6ded4ac5c05ca62265798d26bc"
```

```python
# 웹 데이터 요청하기
resultset = []
with requests.Session() as session:
    session.headers.update({
        "User-Agent" :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) chrome/120.0.0.0 Safari/537.36",
        "Authorization":f"KakaoAK {key}" 
    })

    for i in range(1, 51):
        r = session.get(url, params = {"query":query, "page":i, "size":size, "key":key})

        if r.status_code != 200:
            msg = "[%Error] %s 에러가 발생함" %(r.status_code. r.reason)

            print(msg)
            continue
        print("%d page 데이터 수신 완료 >> %s" % (i, r))

        r.encoding = "utf-8"
        mydict = r.json()
        resultset.extend(mydict['documents'])

print("총 %d 건의 데이터가 수집됨" % len(resultset))
```

```python
# 데이터 저장하기
df = DataFrame(resultset)
df.to_excel("카카오 책 검색(%s)-전체 데이터.xlsx" % query)
```

```python
# 파일 다운로드 함수 정의
def solution(url, target):
    session.headers.update({
        "User-Agent" :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) chrome/120.0.0.0 Safari/537.36"
    })

    try:
        r = session.get(url, stream = True)
        r.encoding = "utf-8"

        with open(target, "wb") as f:
            f.write(r.raw.read())
            print(target,"가 저장되었습니다.")

    except Exception as e:
        print(target, "저장실패", e) 
```

```python
# 반복문을 통해 수집된 결과 내의 표지 이미지 다운로드 받기
import os
import datetime as dt

dirname = dt.datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
os.mkdir(dirname)

for i, v in enumerate(resultset):
    
    file_path = os.path.join(dirname, "%d.png" % i)

    solution(v['thumbnail'], file_path)

```

#04. 네이버 트랜드 데이터 수집

```python
# 라이브러리 참조하기
import requests
import json
import datetime as dt
from pandas import DataFrame
```

```python
# 요청 정보 확인
url = "https://openapi.naver.com/v1/datalab/search"

clientId = "o5C5r88vVM9UIpwRBQTS"
clientSecret = "m35OzmaOy5"

now = dt.datetime.now()
today = now.strftime("%Y-%m-%d")
startDateDelta = now - dt.timedelta(days=365)
startDate = startDateDelta.strftime("%Y-%m-%d")
```

```python
data = {
    "startDate":startDate,
    "endDate" : today,
    "timeUnit":"date",
    "keywordGroups": [
        {
        "groupName":"Benz",
        "keywords":["벤츠", "e클래스", "s클래스", "c클래스", "벤츠 cls"]
    },
    {
        "groupName":"BMW",
        "keywords":["BMW","5시리즈", "x5", "x7", "x3"]
        },
        {
            "groupName":"AUDI",
            "keywords":["아우디", "a6", "e트론", "a7", "q5"]
        }
    ]

}
```

```python
# 웹 데이터 요청하기
with requests.Session() as session:
    session.headers.update({
       "User-Agent" :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) chrome/120.0.0.0 Safari/537.36",
       "X-Naver-Client-Id": clientId,
       "X-Naver-Client-Secret": clientSecret
    })

    r = session.post(url, data = json.dumps(data))

    if r.status_code != 200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    print(r)
 
 
```

```python
# 응답 결과 확인
r.encoding = "utf-8"
mydict = r.json()
mydict["results"]
```

```python
# 응답 결과 재구성
resultset = []

for item in mydict["results"]:
    title = item["title"]
    data = item["data"]

    for i, v in enumerate(data):
        new_data = {"title":title, "period":v["period"], "ratio":v["ratio"]}
        data[i] = new_data

    resultset.extend(data)
resultset 
```

```python
# 최종 데이터셋을 데이터프레임으로 변환
df = DataFrame(resultset)
df
```
