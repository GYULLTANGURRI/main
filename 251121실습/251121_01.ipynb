```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sb
import pandas as pd

# index를 날짜로 변환
bitcoin.index = pd.to_datetime(bitcoin.index)

dpi = 120
font = "Noto Sans Kr"
theme = "dark"

sb.set_theme(style=theme)
plt.rcParams['font.family'] = font
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

width_px = 1000
height_px = 480
fs = (width_px/dpi, height_px/dpi)

fig, ax = plt.subplots(1, 1, figsize=fs, dpi=dpi)

# 시가·종가 추세선
sb.lineplot(data=bitcoin, x=bitcoin.index, y='시가', label='시가', ax=ax)
sb.lineplot(data=bitcoin, x=bitcoin.index, y='종가', label='종가', ax=ax)

# x축 20일 간격
ax.xaxis.set_major_locator(mdates.DayLocator(interval=20))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

plt.grid()
plt.tight_layout()
plt.show()
plt.close()
```
