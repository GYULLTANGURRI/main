üìò#03. WordCloud Í∞ùÏ≤¥Ïùò ÏòµÏÖò

üìù[1]Í∏àÏßÄÏñ¥ ÏÑ§Ï†ï

```python
ignore = set(STOPWORDS)
ignore.add("said")
ignore.add("Alice")
```

üìù[2]ÎßàÏä§ÌÅ¨ Ïù¥ÎØ∏ÏßÄ ÏÑ§Ï†ï

```python
img = Img.open("wordcloud/Ïï®Î¶¨Ïä§Î∞∞Í≤Ω.png")
iarr = np.array(img)
iarr
```

üìù[3]ÏõåÎìú ÌÅ¥ÎùºÏö∞Îìú ÌôïÏù∏

```python
wc = WordCloud(width=600, height=1200,
               max_font_size=150,
               scale=2.0,
               max_words=100,
               stopwords=ignore,
               mask=iarr)
wc.generate(text)
wc.to_file("options.png")
Image(filename='options.png')
```

üìò#03.ÏπºÎùº ÌåîÎ†àÌä∏ ÏÇ¨Ïö©

```python
wc = WordCloud(width=1200, height=800, stopwords = ignore, background_color="#0E0E0E", colormap="plasma")
wc.generate(text)
wc.to_file("colormap.png")
Image(filename='colormap.png')
```
