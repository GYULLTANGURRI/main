# 03.(예제)서울의 고등학교 분포 확인하기

```python
from hossam import load_data
from os import path, mkdir
import folium
```

```python
school =  [37.5613126, 126.9449451]
```

```python
origin = load_data('schools')
origin
```

# 📝2.데이터 전처리

```python
# 1)사용할 필드만 추출
df = origin.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
df
```

```python
df_test = df.copy()
df_test
```

```python
# 2)서울시의 고등학교만 추출

df2 = df.query("학교급구분 == '고등학교' and 소재지도로명주소.str.contains('서울')")
df2
```

# 📝3.데이터 시각화

```python
map5 = folium.Map(location = school, zoom_start = 12)

html = "<fontcolor = 'green' style = 'whitespace: nowrap'> <b>%s </b></font>"

for i in df2.index:
    name = df.loc[i, '학교명']
    lat = df2.loc[i, '위도']
    lng = df2.loc[i, '경도']

    popup_html = folium.Popup(html%name, parse_html=False)
    marker = folium.Marker([lat,lng], popup=popup_html)
    marker.add_to(map5)

    if not path.exists('output'):
        mkdir('output')

        map5.save('output/map5.html')
```
