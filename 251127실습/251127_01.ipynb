```python
from hossam import load_data
from os import path, mkdir
import folium
```

```python
daegangdang = [37.5613126, 126.9449451]
```

```python
# 지도 불러오기
map = folium.Map(location = daegangdang, zoom_start=17)

if not path.exists('output'):
    mkdir('output')

# 지도가 표시된 웹 페이지 파일이 저장됨
# -> 
map.save('output/map.html')
map
```

# 3.지도 객체에 마커 추가

```python
map2 = folium.Map(location=daegangdang, zoom_start = 17)

my_marker = folium.Marker(daegangdang, popup = '이화여대 채플', icon=folium.Icon(color = 'darkred'))

my_marker.add_to(map2)

if not path.exists('output'):
    mkdir('output')

    map2.save('output/map2.html')

map2
```

```python
# 사용자 지정 아이콘, HTML 팝업
map3 = folium.Map(location = daegangdang, zoom_start = 17)

popup_html = folium.Popup("""
    <div style='white-space: nowrap'>
        <h3>이화여자대학교 대강당</h3>
        <p>
            <a href="{google.co.kr/maps/search/이화여자대학교+대강당?entry=ttu&g_ep=EgoyMDI1MTEyMy4xIKXMDSoASAFQAw%3D%3D}" target="_blank">
                구글지도에서 위치 보기
            </a>
        </p>
    </div>
    """,
    parse_html=False
                          
)

icon_img = folium.features.CustomIcon('https://data.hossam.kr/favicon.png', icon_size=(50, 50))

custom_marker = folium.Marker(daegangdang, popup = popup_html, icon=icon_img)
custom_marker.add_to(map3)

if not path.exists('output'):
    mkdir('output')

    map3.save('output/map3.html')
```

# 3)원형 마커(범위지정)

```python
map4 = folium.Map(location = daegangdang, zoom_start = 17)

my_marker = folium.Marker(daegangdang, icon = folium.Icon(color='orange'))

circle_marker = folium.CircleMarker(daegangdang, radius = 100, color ="#000000",  fill_color = '#000000')

circle_marker.add_to(map4)
my_marker.add_to(map4)

if not path.exists('output'):
    mkdir('output')

map4.save('output/map4.html')
```
