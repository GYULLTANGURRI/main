# 고객 데이터와 구매 데이터를 통합하여,연령대별 평균 구매액 분석하기

```python
import pandas as pd
```

```python
# 데이터 프레임 생성

c_df = pd.DataFrame({
    'customer_id':[101,102,103,104,105,106],
    'name':['Alice','Bob','Charlie','David','Eve','Frank'],
    'age':[25,32,45,38,29,41]
})

c_df
```

```python
# 구매내역 불러오기
order_history = load_data("order_history")
order_history
```

```python
# 고객데이터 + 구매데이터 병합
merged= pd.merge(order_history, c_df, on='customer_id', how = 'left')
merged
```

```python
sql join과 같은 방법 - merge
고객데이터 +  구매내역 merge
연령(age group) 만ㄷ르기
연령대별 평균 구매금액 분석
최종 리포트 출력 



고객데이터  + 구매 데이터 
연령(age group) 만드릭
연령대별 평균 구맥므액 분석
최종 리포트 출력

```

```python
# 연령대(age group) 칼럼 만들기
bins = [0,29,39,49,59]
labes = ['20대', '30대', '40대','50대']

merged['age_group'] = pd.cut(merged['age'], bins = bins, labels=labels)
merged
```

```python
결측치처리 = order_history.copy()
mean_price = 결측치처리['price'].mean()
결측치처리['price'] = 결측치처리['price'].fillna(value=mean_price)
결측치처리
```

```python
이상치처리 = 결측치처리.copy()
median_price = 이상치처리['price'].median()
이상치처리.loc[이상치처리['price'] > 1000, 'price'] = median_price
이상치처리
```

```python
병합 = merge(이상치처리, c_df, on = 'customer_id')
병합
```

```python
타입변환 = 병합.astype({'c_df':'str'})
타입변환.describe()
```

# 마케팅팀 신입사원의 첫 번째 미션

```python
from pandas import DataFrame
from pandas import concat
```

```python
vip_customer = load_data('vip_customer')
```

```python
vip_customer.head()
```

```python
# 총 구매액이 100000 이상
cond1 = vip_customer['TotalSpending']>=100000
cond2 = vip_customer['City'] == '서울'

vip_filtered = vip_customer.loc[cond1 & cond2]
vip_filtered
```

```python
# 조건을 충족하는 고객 중 구매액 기준으로 내림차순 정렬
vip_sorted = vip_filtered.sort_values(by = 'TotalSpending', ascending = False)
vip_sorted
```

```python
# 상위 5명의 고객만 추출
top5 =  vip_sorted.head(5)
top5
```

```python
# 보고할 열(Name, City, TotalSpending)만 선택하여 출력
report = top5[['Name','City','TotalSpending']]
report
```

```python
# 전체 코드 한 번에 보기
# 1. 데이터 불러오기
vip = load_data('vip_customer')

# 2. 조건 설정
cond1 = vip['TotalSpending'] >= 100000
cond2 = vip['City'] == '서울'

# 3. 조건을 모두 만족하는 고객 필터링
vip_filtered = vip.loc[cond1 & cond2]

# 4. 구매액 기준 내림차순 정렬
vip_sorted = vip_filtered.sort_values(by='TotalSpending', ascending=False)

# 5. 상위 5명 추출
top5 = vip_sorted.head(5)

# 6. 보고할 컬럼만 선택
report = top5[['Name', 'City', 'TotalSpending']]
report
```
