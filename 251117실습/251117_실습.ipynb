```python
# 문제3-2
pg1 = df2['height_cm'].quantile(0.25)
pg3 = df2['height_cm'].quantile(0.75)
pg_iqr = pg1-pg3
pg_max = pg3 + pg_iqr * 1.5
pg_min = pg1 - pg_iqr * 1.5
print("height_cm에 대한 iqr: %.2f, 하한 이상치 경계:%.2f, 상한 이상치 경계:%.2f "%
      (pg_iqr, pg_min, pg_max))

df2.loc[(df2['height_cm']<pg_min) | (df2['height_cm'] > pg_max)]
```

```python
# IQR 계산
q1 = df2['height_cm'].quantile(0.25)
q3 = df2['height_cm'].quantile(0.75)
iqr = q3 - q1

out_min = q1 - 1.5 * iqr
out_max = q3 + 1.5 * iqr

print("height_cm에 대한 iqr: %.2f, 하한 이상치 경계: %.2f, 상한 이상치 경계: %.2f" %
      (iqr, out_min, out_max))

# 이상치 탐지
df2.loc[(df2['height_cm'] < out_min) | (df2['height_cm'] > out_max)]
```

```python
# sunlight iqr
df3 = df2.copy()
s_h_q3 = df2['sunlight_hours'].quantile(0.25)
s_h_q1 = df2['sunlight_hours'].quantile(0.75)
s_h_iqr = s_h_q3-s_h_q1

sun_min = s_h_q1 - 1.5*s_h_iqr
sun_max = s_h_q3 + 1.5*s_h_iqr
df3.loc[(df3['height_cm'] <out_min) | (df3['height_cm'] > out_max)] = np.nan
print("height_cm에 대한 iqr: %.2f, 하한 이상치 경계: %.2f, 상한 이상치 경계: %2f" %(s_h_iqr, s_h_q1, s_h_q3 ))

df2.loc[(df2['sunlight_hours'] < sun_min) | (df2['sunlight_hours'] > sun_max)]
```
