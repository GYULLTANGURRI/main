# 01.requests 라이브러리

```python
import requests
```

```python
# 웹 데이터 요청하기
with requests.session() as session:
    session.headers.update({
            "User-Agent": "Mozilla/5.0(Winodows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/1"
            })
    
    url = "https://data.hossam.kr/py/sample.txt"
    r = session.get(url)
    if r.status_code != 200:
        msg = "[%] Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    
    print(r)

```

```python
# 응답 결과 확인
r.encoding = "utf-8"
print(r.text)
```

# JSON 데이터 수집

```python
# 새로운 요청 생성하기
with requests.Session() as session:
    session.headers.update({
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64;) Applewebkit/"
        "537.36(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    })

    url = "https://data.hossam.kr/py/sample.json"
    r = session.get(url)

    if r.status_code != 200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    
    print(r)
        
        
        

    
```

# 응답 결과 확인

```python
#JSON을 딕셔너리로 변환

r.encoding = "utf-8"

my_dict = r.json()
my_dict
```

```python
# 필요한 데이터 활용하기
print(my_dict["name"])
print(my_dict["type"])
print(my_dict["img"])
```
