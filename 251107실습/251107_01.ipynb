```python
with requests.Session() as session:
    session.headers.update({
        "User-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"
    })

    url = "https://data.hossam.kr/py/myfood.html"
    r = session.get(url)


    if r.status_code != 200:
        msg = "[%d Error] % 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    
    print(r)

```

```python
r.encoding = "utf-8"

print(type(r.text))
r.text
```

```python
items = soup.select('food-item')

data = []

for item in items:
    names = items.find("h2").textstrip()
    desc = items.find("p").textstrip()
    img_url = items.find["img"]["src"]

    data.append([names, desc, img_url])
    
   
```

```python
with open('myfood.csv', 'w',newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerow(['메일', '설명', '이미지URL'])
    writer.writerow(data)

for img_url, name, desc in data: 
    os.path.name = img_url
    os.path.basename = name
    

```
